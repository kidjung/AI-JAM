```python
import pandas as pd
```

```python
df = pd.read_csv("Kickstarter_projects_Feb19.csv")
df.head()
```

```python
df['main_category'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
successful_df = df[['status']]
failed_df = df[['status']]
successful_cnt = successful_df.loc[successful_df['status'] == 'successful'].count()
failed_cnt = failed_df.loc[failed_df['status'] == 'failed'].count()
print('성공한 데이터 수: ',successful_cnt)
print('실패한 데이터 수: ',failed_cnt)
```

## 성공한 데이터, 실패한 데이터 분리

```python
successful_data = df.loc[df['status'] == 'successful']
failed_data = df.loc[df['status'] == 'failed']
```

```python
successful_data.describe()
```

## 예외 데이터 확인

```python
sns.catplot(data= successful_data, x='main_category', y='goal_usd', kind = 'strip', height = 30)
```

```python
sns.catplot(data= successful_data, x='main_category', y='goal_usd', kind = 'box', height = 30)
```

## USD 80000이상의 값은 예외의 값으로 가정
---

```python
def preprocessing(data):
    data.loc[(data['goal_usd'] > 80000), 'goal_usd'] = 80000
    
    return data
```

## 성공한 펀딩 중 main category 별 목표금액

```python
successful_data = preprocessing(successful_data)
failed_data = preprocessing(failed_data)

sns.catplot(data= successful_data, x='main_category', y='goal_usd', kind='box', height = 10, aspect = 3)
```

```python
sns.catplot(data= successful_data, x='main_category', y='goal_usd', kind='violin', height = 10, aspect = 3, width = 1.2)
```

```python
successful_data_games = successful_data.loc[successful_data['main_category'] == 'games']
successful_data_games.head()
```

##  ex 1) games 카테고리 중 성공한 펀딩 사례들 목표 금액 분포

```python
successful_data_games['sub_category'].value_counts()
```

```python
sns.catplot(data= successful_data_games, x='sub_category', y='goal_usd', kind='violin', height = 10, aspect = 3)
```

## ex 2) crafts 카테고리 중 성공한 펀딩 사례들 목표 금액 분포

```python
successful_data_crafts = successful_data.loc[successful_data['main_category'] == 'crafts']
```

```python
successful_data_crafts['sub_category'].value_counts()
```

```python
sns.catplot(data= successful_data_crafts, x='sub_category', y='goal_usd', kind='violin', height = 10, aspect = 3)
```

# 데이터 분석 결과 및 의견
---

본 문서에서는 킥스타터 데이터 중, 성공한 데이터에서의 각 카테고리 별 목표 금액에 관하여 분석하였음.

데이터 전처리: 
성공한 케이스를 분석하기 위해 실패한 케이스를 제외하고 성공한 케이스만 추출.\
100000달러 이상의 목표치의 경우 데이터의 수가 거의 없기 때문에 (특수한 케이스로 분류) 100000이상의 데이터는 100000달러로 제한함.

데이터 시각화 1: 첫번째로 main category 별 목표 usd를 시각화하여 분석.\
데이터 시각화 2: 특정 main category에서(여기서는 games 카테고리를 분석) sub category 별 목표 usd를 시각화하여 분석.

이를 통해 알 수 있었던 점은  첫째로 각 카테고리별로 예외적으로 아주 높은 목표치를 가지고 펀딩에 성공한 경우도 있지만, 대다수의 경우는 일정한 범위에 밀집되어 있음. 즉, 이 데이터로 어느 정도 신뢰성있는 예측값을 산출 할 수 있을 것으로 기대됨.\
둘째로 적당히 밀집한 범위의 분포치를 가지고 있지 않은 카테고리의 경우, 원인이 데이터의 절대적인 수가 부족한 경우가 대부분임. 이러한 부분은 따로 한 카테고리로 통합시키거나 학습데이터에서 제외시키면 예측값의 정확도가 향상될 것으로 기대됨.
